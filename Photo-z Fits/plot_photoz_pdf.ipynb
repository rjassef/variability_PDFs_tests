```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
tab = Table.read("20random_sources.dat", format='ascii.csv')
```

```python
pdf_file = open("fort.90")
for k in range(len(tab)):
    aux, n = [int(float(ix)) for ix in pdf_file.readline().split()]
    z = np.zeros(n)
    chi2 = np.zeros(n)
    for i in range(n):
        z[i], chi2[i], aux = [float(ix) for ix in pdf_file.readline().split()]
    Pz = np.exp(-0.5*(chi2-chi2.min()))
    norm = np.sum(Pz)
    plt.plot(z, Pz/norm)
    plt.plot([tab['z'][k], tab['z'][k]], plt.ylim(), 'k--')
    plt.show()
```
